```python
import numpy as np
import pandas as pd

def parse_txt(file):
    with open(file, 'r') as f:
        lines = f.readlines()
    lines = [line.strip() for line in lines if line.strip() != '' and line.strip() != '\n']
    return lines

parsed_lines = parse_txt('data/output.txt')
```

```python
uniform_volume_points_occupancies = []
boundary_points_occupancies = []

for i, el in enumerate(parsed_lines):
    if el.startswith("Uniform volume points occupancy"):
        uniform_volume_points_occupancies.append(float(el.split(":")[1].strip()))
    elif el.startswith("Boundary points occupancy"):
        boundary_points_occupancies.append(float(el.split(":")[1].strip()))
```

```python
np.mean(boundary_points_occupancies)
```

```python
import numpy as np
np.mean(uniform_volume_points_occupancies)
```

```python
df = pd.DataFrame({'boundary_points_occupancies': boundary_points_occupancies, 'uniform_volume_points_occupancies': uniform_volume_points_occupancies})
df
```

```python
df.to_csv('f1_occupancies.csv', index=False)
```
